```python
my_list = [('John', 30), ('Jane', 80), ('Olu', 43)]

def name_age_pair():
    current_age = 0
    current_name = ''
    maximum_age = 0
    for name, age in my_list:
        if age > current_age:
            maximum_age = age
            current_name = name
            current_age = age
    return current_age, current_name
name_age_pair()
```
